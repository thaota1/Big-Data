```python
def query(query_string):
    cursorObj.execute(query_string)
    result = cursorObj.fetchall()
    return result    

def showRows(result):
    for row in result:
            print(row)
```

#### 3) Count all tracks with each media type

```python
showRows(query(
    """SELECT media_type.Name,count(track.track_id)
    FROM media_type
    JOIN track ON media_type.media_type_id= track.media_type_id
    group by media_type.media_type_id;
    """))
```

#### 6) Let select the titles of all the songs by Aerosmith in the alphabetical order. Include only the title in the output.

```python
showRows(query("""SELECT songs.title
                    FROM songs, albums,artists
                    WHERE songs.album= albums._id 
                          and albums.artist= artists._id
                          and artists.name = 'Aerosmith'
                    ORDER BY songs.title;
                          """))
```

#### 8) Get the same list as from question 6 but without any duplicates.

```python
showRows(query("""SELECT DISTINCT songs.title
                    FROM songs, albums,artists
                    WHERE songs.album= albums._id 
                          and albums.artist= artists._id
                          and artists.name = 'Aerosmith'
                    ORDER BY songs.title;
                          """))
```

#### 9) Get the count of the songs without duplicates.
Hint: It uses the same keyword as above question but the syntax may not be obvious.

```python
cursorObj.execute("""SELECT count(DISTINCT songs.title)
                    FROM songs, albums,artists
                    WHERE songs.album= albums._id 
                          and albums.artist= artists._id
                          and artists.name = 'Aerosmith'
                    ORDER BY songs.title;
                          """)
result = cursorObj.fetchall()[0]
for rs in result:
    print(rs)
```
