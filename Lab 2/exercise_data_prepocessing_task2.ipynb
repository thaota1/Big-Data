## Handling missing data

```python
train_df.info()
```

```python
names = ['FlightDelayMin','FlightTimeHour','dayOfWeek','hour_of_day']

for name in names:
    train_df[name].replace(0, np.nan, inplace=True)
train_df.head(8)
```

```python
FlightDelayMin_median = train_df['FlightDelayMin'].median()
FlightTimeHour_median = train_df['FlightTimeHour'].median()
dayOfWeek_median = train_df['dayOfWeek'].median()
hour_od_day_median = train_df["hour_of_day"].median()

train_df['FlightDelayMin'].fillna(FlightDelayMin_median , inplace=True)
train_df['FlightTimeHour'].fillna(FlightTimeHour_median, inplace=True)
train_df['dayOfWeek'].fillna(dayOfWeek_median, inplace=True)
train_df["hour_of_day"].fillna(hour_od_day_median, inplace=True)
```
